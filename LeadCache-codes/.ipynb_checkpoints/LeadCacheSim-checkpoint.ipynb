```python
import pandas as pd
import numpy as np
from Lead_cache import Lead_cache
import math
from Generate_network import generate_network_graph
from LRU import Bipartite_LRU
from LFU import Bipartite_LFU
import csv 

users = 10
time = 400
caches = 4
alpha = 0.2
d = 3

Adj = generate_network_graph(users, caches, d)
# generates a random network
```

```python
# Generating the request sequence
#df=pd.read_csv('ratings.csv', sep=',',header=None)
data = pd.read_csv("ratings.dat", sep = "::")
df= pd.DataFrame(data)


#df = df.head(100)
df = df.drop(df.columns[0], axis=1)
df = df.drop(df.columns[1], axis =1)
df.columns = ["file_id", "time_stamp"]

# Dropping all file requests with id larger than the threshold to reduce the library size
threshold=50
df=df[df.file_id<threshold]
df=df.reset_index(drop=True)

library_size = df['file_id'].max()+2
C = math.floor(alpha*library_size)
v=df['file_id']
time = int(np.floor(min(time, len(v)/users)))-1
df=np.array(v)
df = np.array_split(df, users)


# data_points, d = df.shape
# new_id = df.file_id.unique()
# new_id.sort()
# library_size = len(new_id)
# C = math.floor(alpha*library_size)
# new_id = dict(zip(new_id, range(len(new_id))))
# df = df.replace({"file_id": new_id}) 

# shortening the simulation time in case not enough requests are there
#time = int(np.floor(min(time, df.size/users)))-2

#df = np.array_split(df, users)


```

```python
time = min((len(df[0]) - 1, time)) # the request sequence might not have enough data
```

```python
LFU_Hits=[]
LRU_Hits=[]
LeadCache_Hits=[]
```

```python
#LFU_Hits=[]
hit_rates_LFU = pd.DataFrame(Bipartite_LFU(df, Adj, time, library_size, C))
#np.sum(hit_rates_LFU)
LFU_Hits.append(np.sum(hit_rates_LFU))
#hit_rates_LFU.to_csv('LFU.csv',index=False)
```

```python
#LRU_Hits=[]
hit_rates_LRU = pd.DataFrame(Bipartite_LRU(df, Adj, time, library_size, C))
np.sum(hit_rates_LRU)
LRU_Hits.append(np.sum(hit_rates_LRU))
#hit_rates_LRU.to_csv('LRU.csv',index=False)  
```

```python
LeadCache_Hits=[]
hit_rates_Lead_cache = pd.DataFrame(Lead_cache(df, Adj, time, library_size, C, d))
LeadCache_Hits.append(np.sum(hit_rates_Lead_cache))
#hit_rates_Lead_cache.to_csv('Lead_cache.csv',index=False) 
```
